# Exploration du dataset "wine"

## Importation des bibliothèques nécessaires dans notre code

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
```

# Début de l'analyse exploratoire

## Téléchargement du dataset depuis UCI (le site dont provient le dataset)

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
columns = [
    "age", "workclass", "fnlwgt", "education", "education_num", 
    "marital_status", "occupation", "relationship", "race", 
    "sex", "capital_gain", "capital_loss", "hours_per_week", 
    "native_country", "income"
]
data = pd.read_csv(url, header=None, names=columns, na_values=" ?", skipinitialspace=True)
```

## Nettoyage des données et préparation pour les classificateurs

```python
print("Avant nettoyage :")
print(data.isnull().sum())
data.dropna(inplace=True)
print("\nAprès nettoyage :")
print(data.isnull().sum())

#Conversion des variables catégoriques en numériques (afin de pouvoir les utiliser avec les classificateurs)
for col in data.select_dtypes(include="object").columns:
    le = LabelEncoder()
    data[col] = le.fit_transform(data[col])

#Séparation des paramètres et de la valeur cible
X = data.drop(columns=["income"])
y = data["income"]

#Standardisation des données
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

## Test des modèles de classification

```python
#Séparation en ensembles d'entraînement et de test pour les modèles
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.3, random_state=42)

#Initialisation des modèles de classificateurs
models = {
    "SVM": SVC(probability=True, random_state=42),
    "Logistic Regression": LogisticRegression(random_state=42),
    "Random Forest": RandomForestClassifier(random_state=42),
    "Gradient Boosting": GradientBoostingClassifier(random_state=42),
    "Gaussian Naive Bayes": GaussianNB()
}
#On ne peut pas utiliser Multinomial à cause des valeurs négatives dans le datset, nous gardons par conséquent seulement GaussianNB
#Test des modèles
results = []
for name, model in models.items():
    print(f"--- {name} ---")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"Accuracy: {accuracy:.2f}")
    print(classification_report(y_test, y_pred))
    results.append({"Model": name, "Accuracy": accuracy})

#Résultats des modèles testés
results_df = pd.DataFrame(results).sort_values(by="Accuracy", ascending=False)
print("Résumé des performances des modèles :")
print(results_df)
```

```python
#Visualisation des performances obtenu grâce aux tests
plt.figure(figsize=(8, 6))
sns.barplot(x="Accuracy", y="Model", data=results_df, palette="viridis")
plt.title("Comparaison des modèles selon leur précision")
plt.xlabel("Accuracy")
plt.ylabel("Model")
plt.xlim(0, 1)

#Enregistrement du graphique
plt.savefig("comparaison_modeles_precision.png", dpi=300, bbox_inches="tight")

plt.show()
```

# Nous choisirons donc Gradient Boosting pour la suite de l'exploration de ce dataset

## Réentraînement du modèle et importance des caractéristiques

```python
gb_model = GradientBoostingClassifier(random_state=42)
gb_model.fit(X_train, y_train)
y_pred_gb = gb_model.predict(X_test)
print("Rapport de classification :")
print(classification_report(y_test, y_pred_gb))

#Importances des caractéristiques
feature_importances = pd.DataFrame({
    "Feature": columns[:-1],
    "Importance": gb_model.feature_importances_
}).sort_values(by="Importance", ascending=False)

print("\nImportances des caractéristiques :")
print(feature_importances)

plt.figure(figsize=(10, 6))
plt.barh(feature_importances["Feature"], feature_importances["Importance"], color="coral")
plt.gca().invert_yaxis()
plt.title("Importances des caractéristiques (Gradient Boosting)")
plt.xlabel("Importance")
plt.ylabel("Caractéristiques")
plt.savefig("importancecaractéristiques.png", dpi=300, bbox_inches="tight")

plt.show()

#Génération et affichage de la matrice de confusion
conf_matrix = confusion_matrix(y_test, y_pred_gb)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Matrice de confusion - Gradient Boosting")
plt.xlabel("Prédictions")
plt.ylabel("Valeurs réelles")
plt.savefig("matriceconfusionhyperpara1.png", dpi=300, bbox_inches="tight")
plt.show()
```

### Optimisation des hyperparamètres avec GridSearchCV

```python
param_grid = {
    "n_estimators": [50, 100, 200],
    "max_depth": [3, 5, 10],
    "learning_rate": [0.01, 0.1, 0.2],
    "subsample": [0.8, 1.0],
    "min_samples_split": [2, 5],
    "min_samples_leaf": [1, 2]
}
grid_search = GridSearchCV(
    GradientBoostingClassifier(random_state=42),
    param_grid,
    cv=3,
    scoring="accuracy",
    verbose=2,
    n_jobs=-1
)
grid_search.fit(X_train, y_train)
print("\nMeilleurs paramètres :", grid_search.best_params_)

best_gb = grid_search.best_estimator_
y_pred_optimized = best_gb.predict(X_test)
optimized_accuracy = accuracy_score(y_test, y_pred_optimized)
print(f"\nAccuracy après optimisation : {optimized_accuracy:.2f}")
```

### Vérification de l'équilibre des classes et pondération de celles-ci

```python
#Vérification de l'équilibre
class_distribution = y_train.value_counts()
print("Distribution des classes dans l'ensemble d'entraînement :")
print(class_distribution)

#Calcul du ratio entre les classes
class_weights = {cls: len(y_train) / (len(class_distribution) * count) for cls, count in class_distribution.items()}
print("\nPondération calculée pour chaque classe :")
print(class_weights)

#Réentraînement du modèle Gradient Boosting avec pondération des classes
gb_weighted = GradientBoostingClassifier(
    random_state=42,
    **grid_search.best_params_  #Utilisation des meilleurs hyperparamètres trouvés précédemment
)
#Application de la pondération aux échantillons
sample_weights = y_train.map(class_weights).values
gb_weighted.fit(X_train, y_train, sample_weight=sample_weights)

#Prédictions et évaluation
y_pred_weighted = gb_weighted.predict(X_test)
accuracy_weighted = accuracy_score(y_test, y_pred_weighted)
print(f"\nAccuracy avec pondération des classes : {accuracy_weighted:.2f}")
print("Rapport de classification avec pondération :")
print(classification_report(y_test, y_pred_weighted))

#Génération et affichage de la matrice de confusion après pondération
conf_matrix_weighted = confusion_matrix(y_test, y_pred_weighted)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_weighted, annot=True, fmt="d", cmap="Purples", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Matrice de confusion - Gradient Boosting (avec pondération)")
plt.xlabel("Prédictions")
plt.ylabel("Valeurs réelles")
plt.savefig("matriceconfusionapresponderation.png", dpi=300, bbox_inches="tight")

plt.show()
```

### Utilisation de poids sur les caractéristiques afin de voir si cela affecte les résultats

#### Ajustement manuel des hyperparamètres en fonction des résultats précédents
#### Ici, on suppose que "max_depth", "learning_rate" et "n_estimators" sont les hyperparamètres les plus cohérents

```python
optimized_params = grid_search.best_params_

#Renforcement de l'importance des paramètres clés
optimized_params["learning_rate"] *= 1.2  # Augmenter légèrement le taux d'apprentissage
optimized_params["max_depth"] += 1       # Rendre l'arbre un peu plus complexe
optimized_params["n_estimators"] += 50   # Ajouter plus d'estimateurs pour un meilleur ajustement

print("Nouveaux hyperparamètres après ajustement :", optimized_params)

#Réentraîner le modèle avec les nouveaux hyperparamètres sélectionnés
final_gb = GradientBoostingClassifier(random_state=42, **optimized_params)
final_gb.fit(X_train, y_train)

#Évaluation du modèle final
y_pred_final = final_gb.predict(X_test)
final_accuracy = accuracy_score(y_test, y_pred_final)
print(f"\nAccuracy après ajustement des hyperparamètres : {final_accuracy:.2f}")
print("Rapport de classification :")
print(classification_report(y_test, y_pred_final))

#Génération et affichage de la matrice de confusion après ajustement des hyperparamètres
conf_matrix_adjusted = confusion_matrix(y_test, y_pred_final)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix_adjusted, annot=True, fmt="d", cmap="Greens", xticklabels=le.classes_, yticklabels=le.classes_)
plt.title("Matrice de confusion - Gradient Boosting (hyperparamètres ajustés)")
plt.xlabel("Prédictions")
plt.ylabel("Valeurs réelles")
plt.savefig("matriceconfusionhyperparaajutes.png", dpi=300, bbox_inches="tight")

plt.show()
```
